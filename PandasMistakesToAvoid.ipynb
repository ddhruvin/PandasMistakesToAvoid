```python
#w/o @
min_year = 1980
min_time = 10
df.query('Year < ' + str(min_year) + 'and Time >' + str(min_time))
```

#inplace = True will be removed in future versions, so try not to use it

```python
#w/ inplace = True

"""setting inplace = True will overwrite df with the changes
If inplace = False: returns a copy where the operation is done."""
df.fillna(0, inplace=True)
df.head()
```

#Not setting suffixes when merging two dataframes on a common column

```python
#Instead use the validate parameter in .merge function
df_merged = df.merge(new_df, on = 'Athlete', suffixes=('_old', '_new'), validate='m:m')

# validate options:
# "one_to_one" or "1:1": check if merge keys are unique in both datasets.
# "one_to_many" or "1:m": check if merge keys are unique in left dataset.
# "many_to_one" or "m:1": check if merge keys are unique in right dataset. \
# "many_to_many" or "m:m": allowed, but does not result in checks.

#It will throw a MergeError if the validation fails (here at 1:m)
```

#Concatenating 2 dfs can create duplicate cols, remove them by

```python
df_concat = pd.concat([df, new_df], axis=1)
df_concat.head()
```

```python
#Set a flag using set_flags method in the .concat function
df_concat = pd.concat([df, new_df], axis=1) \
.set_flags(allows_duplicate_labels=False)
df_concat.head()
```

```python
#Solve the duplication problem by removing them
df_concat= df_concat.loc[: , ~df_concat.columns.duplicated()].copy()
df_concat.head()
```
